```python
# NEWTON'S LAW OF COOLING
import math
def Newtons_law_of_cooling(t,Tt,Ts,To):
    k = -(1/t) * math.log((Tt-Ts)/(To-Ts))          # k is the proportionality constant
    T = Ts+(To-Ts) * math.exp(-k*t)                 # T is the temperature at time t
    return T
t = 11                                          # time at which the temperature we need
Tt = 270                                        # Tt is the temperature of the body at a particular time
Ts = 300                                        # Ts is the temperature of the surrounding
To = 29                                         # To is the initial temperature of the body
Newtons_law_of_cooling(t,Tt,Ts,To)
```
